```python
import json
import spacy
import copy
import re
import time
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

start = time.time()
custom_stopwords = set(["https", "http", "www", "html", "com", "st", "rd", "nd", "th"])

# Load the spacy model
nlp = spacy.load("en_core_web_sm")

# Custom Preprocessor and Tokenizer
def custom_preprocessor(text):
    text = text.replace('\n', ' ').replace('\t', ' ')
    text = text.replace('\\n', ' ').replace('\\t', ' ')
    text = re.sub(r"[^a-zA-Z\s]", " ", text)
    text = re.sub(r"\d+", " ", text)  # Remove digits
    return text.lower()

def custom_tokenizer(text):
    doc = nlp(text)
    return [token.lemma_ for token in doc if not token.is_stop and not token.is_punct and token.lemma_.strip() and token.lemma_.lower() not in custom_stopwords and len(token.lemma_) > 1]

# LDA parameters
n_topics = 20  # Adjust the number of topics
n_top_words = 6  # Top words per topic

for fold_num, (train_idx, test_idx) in enumerate(folds):
    fold_data = copy.deepcopy(data)
    all_docs_train = []
    doc_positions_train = []  
    all_docs_test = []
    doc_positions_test = []  
    
    # Splitting data into train and test sets for the current fold
    for query_id, entry in data.items():
        for idx, doc in enumerate(entry["top_docs"]):
            if int(query_id) in train_idx:
                all_docs_train.append(doc["0"])
                doc_positions_train.append((query_id, idx))
            elif int(query_id) in test_idx:
                all_docs_test.append(doc["0"])
                doc_positions_test.append((query_id, idx))

    print(f"Documents extracted for fold{fold_num}, start count vectorizer ...")
    print(f"Size of train for fold{fold_num} is {len(all_docs_train)}")
    print(f"Size of test for fold{fold_num} is {len(all_docs_test)}")

    # Vectorizing training data with TF-IDF for preprocessing
    vectorizer = CountVectorizer(
        preprocessor=custom_preprocessor,
        tokenizer=custom_tokenizer,
        lowercase=False  # Already done in preprocessor
    )

    tfidf_train = vectorizer.fit_transform(all_docs_train)
    tfidf_test = vectorizer.transform(all_docs_test)
    feature_names = vectorizer.get_feature_names_out()
    print("TFIDF finished, start LDA modeling ...")
    
    # Apply LDA on training data
    lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
    lda.fit(tfidf_train)

    # Extract the top words for each topic
    topic_words = []
    for topic_idx, topic in enumerate(lda.components_):
        top_word_indices = topic.argsort()[::-1][:n_top_words]
        top_words = [feature_names[i] for i in top_word_indices]
        topic_words.append(top_words)
    
    print("LDA topic modeling completed, start document topic assignment ...")

    # Step 2: Represent each document with top words from top 3 topics
    for i, (query_id, doc_idx) in enumerate(doc_positions_train):
        if i % 100 == 0: 
            print("Finished 100 docs reps (train)")
        
        # Get the topic distribution for the current document
        doc_topic_dist = lda.transform(tfidf_train[i])[0]
        top_topic_indices = np.argsort(doc_topic_dist)[::-1][:3]  # Top 3 topics

        # Collect the top words from each of the top topics
        top_words = []
        for topic_idx in top_topic_indices:
            top_words.extend(topic_words[topic_idx])
        
        top_words = top_words[:18]  # Limit to 18 words

        # Update the document representation
        fold_data[query_id]["top_docs"][doc_idx] = " ".join(top_words)

    # Apply the same logic for the test set
    for i, (query_id, doc_idx) in enumerate(doc_positions_test):
        if i % 100 == 0:
            print("Finished 100 docs reps (test)")
        
        # Get the topic distribution for the current document
        doc_topic_dist = lda.transform(tfidf_test[i])[0]
        top_topic_indices = np.argsort(doc_topic_dist)[::-1][:3]  # Top 3 topics

        # Collect the top words from each of the top topics
        top_words = []
        for topic_idx in top_topic_indices:
            top_words.extend(topic_words[topic_idx])
        
        top_words = top_words[:18]  # Limit to 18 words

        # Update the document representation
        fold_data[query_id]["top_docs"][doc_idx] = " ".join(top_words)

    # Save modified structure after processing the fold
    with open(f'FOLD{fold_num}_training_top20_qulac_en_LDA_rep.json', 'w', encoding='utf-8') as f_out:
        json.dump(fold_data, f_out, ensure_ascii=False, indent=2)
    print(f"Fold {fold_num} data saved")

print(f"Total time taken: {time.time() - start} seconds")
```
